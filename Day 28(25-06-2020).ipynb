```python
# importing the library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## loading the data set
from sklearn.datasets import load_boston
boston=load_boston()
```

```python
boston_df=pd.DataFrame(boston.data,columns=boston.feature_names)
boston_df.head()
```

```python
boston_df["MEDV"]=boston.target
boston_df.head()
```

```python
## to find the correlation we use corr()
boston_df.corr()
```

```python
## visualizing the correlation
plt.figure(figsize=(12,12))
sns.heatmap(boston_df.corr(),annot=True,cmap="YlGnBu")
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(boston_df.corr()[boston_df.corr()>0.7],annot=True,cmap="YlGnBu")
```

```python
correlation=boston_df.corr()
cor = correlation[correlation>0.7]
plt.figure(figsize=(12,12))
sns.heatmap(cor,cmap="Greens",annot=True)
```

```python
## All the coloumns as input 
## drop the columns which are highly correlated and need to compare the perfromance 
```

```python
## spliting into training and testing , building the model in training and test model on testing data
```

```python
## split the data input and output
X=boston_df.drop(["MEDV"],axis=1)
y=boston_df.MEDV
```

```python
# Divide the data to train and test 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
# building the model on training data --> X_train and y_train
# sklearn lib: import,instantiate,fit 
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)
lm.coef_
```

```python
coefficients=pd.DataFrame([X_train.columns,lm.coef_]).T
coefficients
```

```python
y_train_pred=lm.predict(X_train)
y_train_pred
```

```python
## we need to compare the actual values(y_train) and predicted values(y_tarin_pred)
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

print("R^2 :",r2_score(y_train,y_train_pred))

print("Adjusted R^2 :",1-(1-r2_score(y_train,y_train_pred))*(len(X_train)-1)/
      (len(X_train)-X_train.shape[1]-1))
print("MAE :",mean_absolute_error(y_train,y_train_pred))

print("MSE :",mean_squared_error(y_train,y_train_pred))

print("RMSE :",np.sqrt(mean_squared_error(y_train,y_train_pred)))
```

```python
numerator = (1-r2_score(y_train_pred,y_train))*(len(X_train)-1)
denominator = (len(X_train)-X_train.shape[1]-1)

print("Adjusted R^2",1- (numerator/denominator))
```

```python
## regplot(y_train --> actual values adn y_train_pred )
sns.regplot(y_train,y_train_pred)
```

```python
# visualize the differences between the actual and the predicted values

plt.scatter(y_train,y_train_pred)
plt.xlabel("Price")
plt.ylabel("Model Predicted Price")
```

```python
## y_train,y_train_pred==Errors
plt.scatter(y_train_pred,y_train-y_train_pred)
plt.xlabel("Predicted values")
plt.ylabel("Residuals")
```

```python
## Test the model on testing data
y_test_pred=lm.predict(X_test) # test data-- unseen data
y_test_pred
```

```python
## we need to compare the actual values(y_test) and predicted values(y_test_pred)
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

print("R^2 :",r2_score(y_test,y_test_pred))

print("Adjusted R^2 :",1-(1-r2_score(y_test,y_test_pred))*(len(X_test)-1)/
      (len(X_test)-X_test.shape[1]-1))

print("MAE :",mean_absolute_error(y_test,y_test_pred))

print("MSE :",mean_squared_error(y_test,y_test_pred))

print("RMSE :",np.sqrt(mean_squared_error(y_test,y_test_pred)))
```
